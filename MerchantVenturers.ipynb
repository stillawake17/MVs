```python
import pandas as pd

# Load the Excel file
data = pd.read_excel("C:\\Users\\josti\\OneDrive\\Documents\\00 - Ephemeral Boutique\\0 - data analysis\\Merchant Venturers\\members.xlsx")

# Display the first few rows of the data
data.head(20)
```

```python
import networkx as nx

# Create an empty network
G = nx.Graph()

# Add nodes and edges based on the data
for index, row in data.iterrows():
    person = row[0]
    G.add_node(person)  # Add the person as a node
    
    for link in row[1:].dropna():
        G.add_node(link)  # Add the organization/person as a node
        G.add_edge(person, link)  # Add an edge between the person and the organization/person

# Return basic information about the network
G_info = nx.info(G)
G_info
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(13, 13))

# Draw the network
pos = nx.spring_layout(G, seed=42)  # Positioning of nodes
nx.draw(G, pos, with_labels=True, node_size=600, node_color="skyblue", font_size=8, font_weight="bold")

plt.title("Network Visualization")
plt.show()
```

```python
import pandas as pd
import networkx as nx
from pyvis.network import Network

# Load data from Excel
#data = pd.read_excel("path_to_your_file/members.xlsx")

# Load the Excel file
data = pd.read_excel("C:\\Users\\josti\\OneDrive\\Documents\\00 - Ephemeral Boutique\\0 - data analysis\\Merchant Venturers\\members.xlsx")


# Create a NetworkX graph
G = nx.Graph()

# Populate graph from the data
for index, row in data.iterrows():
    person = row[0]
    G.add_node(person)
    
    for link in row[1:].dropna():
        G.add_node(link)
        G.add_edge(person, link)

# Create an interactive network using pyvis
net = Network(notebook=True, height="750px", width="100%")
net.from_nx(G)

# Customize nodes for better visualization
for node in net.nodes:
    node["title"] = node["label"]
    node["value"] = 5

# Save and show the network
net.show("network.html")
```
